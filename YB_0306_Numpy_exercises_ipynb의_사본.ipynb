#### 25. What are the result of the following expressions? (★☆☆)
```python
np.array(0) / np.array(0)
np.array(0) // np.array(0)
np.array([np.nan]).astype(int).astype(float)
```



> 인용구 추가



````python
import numpy as np
print(np.array(0) / np.array(0))
print(np.array(0) // np.array(0))
print(np.array([np.nan]).astype(int).astype(float))

# 0/0은 정의되지 않는 값이므로, 숫자가 아닌 값(오류값) not a number을 의미하는 non이 출력된다.
# '//'는 'floor division'을 의미하며, '몫'만 구하는 나눗셈이다. 0/0의 몫인 0이 출력된다.
# 'nan'을 astype(int)을 통해 데이터값을 최소 정수값으로 바꾸었다가, float을 통해 그 값을 실수로 표현한다.
````

#### 26. How to find common values between two arrays? (★☆☆)
```python
Z1 = np.random.randint(0,10,10)
Z2 = np.random.randint(0,10,10)
```
Hint : use `numpy.random.randint(start, end, size)`

````python
Z1 = np.random.randint(0,10,10)
Z2 = np.random.randint(0,10,10)
print(np.intersect1d(Z1,Z2))

# 0부터 9까지의 10개의 수를 생성한다. intersect1d함수를 통해, Z1,Z2 둘 다에 있는 값을 찾는다. 두 함수 Z1,Z2의 교집합 값이 4,5,8이다. 매번 시행할 때마다 다른 값이 나온다.
````

#### 27. Is the following expressions true? (★☆☆)
```python
np.sqrt(-1) == np.emath.sqrt(-1)
```

False. nan!=1j


````python
import numpy as np
print(np.sqrt(-1) == np.emath.sqrt(-1))

#파이썬에서는 emath를 활용하면 복소수값을 반환된다. np.sqrt(-1)은 정의되지 않은 값으로 nan'값이 출력되고, np.emath.sqrt(-1)은 '1j'값(j는 허수단위)이 출력된다.
두 값은 같지 않으므로 'False'가 된다.
````
