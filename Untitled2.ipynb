# DataFrame과 리스트, 딕셔너리, 넘파이 ndarray 상호 변환

1. ndarray/리스트/딕셔너리 -> DataFrame 변환

DataFrame은 행(row)와 열(column)을 가진 2차원 데이터 구조

(1) 1차원 데이터 -> DataFrame

: 리스트나 ndarray가 1차원이면 컬럼이 1개 필요

(데이터: [1,2,3] -> DataFrame 컬럼 하나 생성)

(2) 2차원 데이터 -> DataFrame

데이터가 2차원이면 열 개수만큼 컬럼 이름 필요

(데이터: [[1,2,3],

 [11,12,13]]

col1 col2 col3)

(3) 딕셔너리 -> DataFrame

딕셔너리를 DataFrame으로 만들 때

- key: 칼럼 이름

- value: 컬럼 데이터

{'col1':[1,11],

'col2':[2,22],

'col3':[3,33]}

-> 각 key가 DataFrame의 열이 됨

2. DataFrame -> ndarray/리스트/딕셔너리 변환

(1) ndarray 변환

DataFrame의 데이터를 numpy 배열로 변환

-> values 사용

- DataFrame -> ndarray 변환은 머신러닝에서 많이 사용

왜냐하면 많은 ML 라이브러리가 numpy 배열을 입력으로 사용하기 때문

(2) 리스트 변환

DataFrame -> ndarray -> list 변환

-> 리스트 형태 데이터로 바뀜

(3) 딕셔너리 변환

DataFrame -> dict

-> 킬럼을 key로 가지는 딕셔너리 생성

# DataFrame 컬럼 생성/수정

DataFrame에서는 새로운 컬럼을 쉽게 만들 수 있음

(1) 새로운 컬럼 추가

DataFrame['새컬럼'] = 값

-> 전체 행에 값이 들어감

(2) 기존 컬럼으로 새 컬럼 생성

기존 컬럼을 이용해 새 변수(feature) 만들 수 있음

( Age x 10

가족 수 계산)

-> z컬럼 연산 가능

(3) 컬럼 값 업데이트

컬럼 전체 값을 한 번에 변경 가능

(컬럼 = 컬럼 + 100)

-> 모든 행 값이 동시에 수정
